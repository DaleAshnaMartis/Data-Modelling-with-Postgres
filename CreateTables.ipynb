```python
import psycopg2
from sql_queries import create_table_queries, drop_table_queries
from configparser import ConfigParser
```

```python
def get_connection_byconfig(filepath,section):
    if (len(filepath)>0 and len(section)>0):
# creates instance of the Config Parser class
        parser=ConfigParser()
# reads the configuration file
        parser.read(filepath)
        diction={}
# if the configuration file has the passed section name
        if (parser.has_section(section)):
# this creates each line in the config file has a separate list. Example ('host','localhost')
            config_params=parser.items(section)
# convert list to python dictionary object
# create empty dictionary
          
            for c in config_params:
                key=c[0]
                value=c[1]
                diction[key]=value
        return diction

```

```python
def create_database():
    # connect to default database
    d=get_connection_byconfig('Config_ini','postgresql')
    conn = psycopg2.connect(host=d['host'],dbname='postgres', user=d['user'], password=d['password'],port=d['port'])
    #conn = psycopg2.connect(host=d['host'], user=d['user'], password=d['password'],port=d['port'])
    print(d['host'],d['user'],d['password'],d['port'])
    

    conn.set_session(autocommit=True)
    cur = conn.cursor()

    # create sparkify database with UTF8 encoding
    
    cur.execute("DROP DATABASE IF EXISTS sparkifydb")
    cur.execute("CREATE DATABASE sparkifydb WITH ENCODING 'utf8' TEMPLATE template0")

    # close connection to default database
    conn.close()
    
    # 
    conn = psycopg2.connect(host=d['host'],dbname='sparkifydb', user=d['user'], password=d['password'],port=d['port'])
    cur = conn.cursor()
    
    return cur, conn
```

```python
def drop_tables(cur, conn):
    for query in drop_table_queries:
        cur.execute(query)
        conn.commit()


def create_tables(cur, conn):
    for query in create_table_queries:
        cur.execute(query)
        conn.commit()
```

```python
def main():
    cur, conn = create_database()
    drop_tables(cur, conn)
    create_tables(cur, conn)

    
    conn.close()
```

```python
if __name__ == "__main__":
    main()
```
